```python
data = pd.read_csv('census.csv')
```

```python
data.head()
```

```python
data = data.applymap(lambda x: x.strip().replace(" ", "") if type(x) == str else x)
data.rename(columns=lambda x: x.strip(), inplace=True)
data.rename(columns=lambda x: x.strip().replace("-", "_"), inplace=True)
```
